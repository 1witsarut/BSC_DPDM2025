# Data

```python
from sklearn.datasets import load_digits

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# You can access the data and target with:
# X, y = mnist.data, mnist.target

print(f"Data shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Number of class: {len(digits.target_names)}")
print(f"First 5 target labels: {y[:5]}")
```

```python
import matplotlib.pyplot as plt
first_image_data = X[0] # Get the first image data from X
# The line 'first_image = y[0]' was causing the error as it overwrites the image data with a single label
image_reshape = first_image_data.reshape(8, 8) # Reshape the image data
plt.imshow(image_reshape, cmap='gray') # Display the reshaped image
```

```python
import pandas as pd

# Create a DataFrame from the feature data X
df = pd.DataFrame(X)

# Add the target labels y as a new column
df['target'] = y

print("Pandas DataFrame created successfully.")
print("First 5 rows of the DataFrame:")
print(df.head())
```

```python
df.head()
```

```python
X_train = X[:1700]
y_train = y[:1700]
X_test = X[1700:]
y_test = y[1700:]
```

## Model Construction
- import
- define(hyper-paramiter)
- train(fitting)
- evalue

## import

```python
from sklearn.tree import DecisionTreeClassifier
```

## Define

```python
tree0 = DecisionTreeClassifier(random_state=6)
```

## training

```python
tree0.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
# Convert class_names to strings
string_class_names = [str(name) for name in digits.target_names]
plot_tree(tree0, filled=True, feature_names=digits.feature_names, class_names=string_class_names)
plt.show()
```

## Evaluate

```python
y_predict = tree0.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```
