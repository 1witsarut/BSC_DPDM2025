# Data

```python
from sklearn.datasets import load_digits

# Load the MNIST dataset
digits = load_digits()
X, y = digits.data, digits.target

# You can access the data and target with:
# X, y = mnist.data, mnist.target

print(f"Data shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Number of class: {len(digits.target_names)}")
print(f"First 5 target labels: {y[:5]}")
```

```python
import matplotlib.pyplot as plt
first_image_data = X[0] # Get the first image data from X
# The line 'first_image = y[0]' was causing the error as it overwrites the image data with a single label
image_reshape = first_image_data.reshape(8, 8) # Reshape the image data
plt.imshow(image_reshape, cmap='gray') # Display the reshaped image
```

```python
import pandas as pd

# Create a DataFrame from the feature data X
df = pd.DataFrame(X)

# Add the target labels y as a new column
df['target'] = y

print("Pandas DataFrame created successfully.")
print("First 5 rows of the DataFrame:")
print(df.head())
```
