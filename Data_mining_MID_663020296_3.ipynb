```python
import pandas as pd
import glob
import os

# 1. กำหนด Path หลักของโฟลเดอร์ Input (อ้างอิงจากโฟลเดอร์ที่คุณมี)
base_path = '/content/drive/MyDrive/BSC_DPDM2025/mid/Input - (Operation Daily rain from API_ 2025)'

# 2. ดึงรายชื่อไฟล์ .csv ทั้งหมดจากทุกโฟลเดอร์ย่อย
all_files = glob.glob(os.path.join(base_path, "**/*.csv"), recursive=True)

df_list = []

print(f"กำลังเริ่มอ่านไฟล์ทั้งหมด {len(all_files)} ไฟล์...")

for f in all_files:
    # อ่านไฟล์ (ใช้ error_bad_lines=False หากบางไฟล์มีโครงสร้างแถวผิดปกติ)
    temp_df = pd.read_csv(f)

    # เลือกเฉพาะคอลัมน์ที่จำเป็น:
    # - rainfall_datetime: วันที่
    # - rainfall_value: ค่าน้ำฝน
    # - tele_station_oldcode: รหัสสถานีแบบตัวอักษร (เพื่อเอาไปเทียบกับไฟล์ Station Master และ Historical)
    required_cols = ['rainfall_datetime', 'rainfall_value', 'tele_station_oldcode']

    # ตรวจสอบว่าไฟล์มีคอลัมน์ครบตามต้องการไหม
    if all(col in temp_df.columns for col in required_cols):
        # เก็บเฉพาะคอลัมน์ที่เลือก และลบแถวที่รหัสสถานีเป็นค่าว่าง
        clean_temp = temp_df[required_cols].dropna(subset=['tele_station_oldcode'])
        df_list.append(clean_temp)

# 3. รวมข้อมูลทั้งหมดเข้าด้วยกัน
df_input = pd.concat(df_list, ignore_index=True)

# 4. แปลงคอลัมน์วันที่ให้เป็นรูปแบบ Datetime เพื่อใช้คำนวณในข้อถัดไป
df_input['rainfall_datetime'] = pd.to_datetime(df_input['rainfall_datetime'])

# --- แสดงผลลัพธ์ข้อ 1 ---
print("\n--- สรุปโครงสร้างข้อมูลข้อ 1 ---")
print(df_input.info())
display(df_input.head())
```

อธิบายคอลัม

rainfall_datetime:ความหมาย: วันและเวลาที่ทำการบันทึกข้อมูลการวัดน้ำฝน

rainfall_value:ความหมาย: ปริมาณน้ำฝนที่วัดได้จริง ณ สถานีนั้นๆ มีหน่วยเป็น มิลลิเมตร (mm)

tele_station_oldcode:ความหมาย: รหัสประจำสถานีตรวจวัดอากาศในรูปแบบตัวอักษร (เช่น ACRU, BBHN)

ข้อ2

```python
# --- ข้อ 2: กรองข้อมูลสถานี ---
# 1. อ่านไฟล์สถานี Master
path_station_model = '/content/drive/MyDrive/BSC_DPDM2025/mid/Input - (Operation Daily rain from API_ 2025)/Stations in the model_HII.csv'
df_stations_master = pd.read_csv(path_station_model)

# 2. กรองเฉพาะสถานีที่มี Node ตรงกับ tele_station_oldcode ในข้อมูลหลัก
df_filtered = df_input[df_input['tele_station_oldcode'].isin(df_stations_master['Node'])].copy()

print(f"จำนวนสถานีก่อนกรอง: {df_input['tele_station_oldcode'].nunique()}")
print(f"จำนวนสถานีหลังกรอง: {df_filtered['tele_station_oldcode'].nunique()}")
```

ข้อ3

```python
# --- ข้อ 3: ตรวจสอบข้อมูลสูญหาย ---

# 1. เช็ค missing value ในคอลัมน์น้ำฝน (แบบที่เรียนมา: .isnull().sum())
print("จำนวน Missing ในคอลัมน์น้ำฝน:")
print(df_filtered['rainfall_value'].isnull().sum())

# 2. เช็ค 'วันที่' ที่หายไป (Record ที่หายไป)
expected_days = 184
df_filtered['date_only'] = df_filtered['rainfall_datetime'].dt.date

# สร้างตารางสรุปการส่งข้อมูลรายสถานี
summary = df_filtered.groupby('tele_station_oldcode')['date_only'].nunique().reset_index()
summary.columns = ['Node', 'actual_days']

# คำนวณจำนวนวันที่หายไปและสัดส่วนเปอร์เซ็นต์
summary['missing_count'] = expected_days - summary['actual_days']
summary['missing_ratio'] = (summary['missing_count'] / expected_days) * 100

print("\n--- สรุปข้อมูลสูญหายรายสถานี (ข้อ 3) ---")
display(summary.head())
```

ข้อ4

```python
# 1. หารายชื่อสถานีที่ต้องถูก Drop (ตามเกณฑ์ missing_ratio >= 20)
drop_list = summary[summary['missing_ratio'] >= 20]['Node'].tolist()

# 2. กรองข้อมูลให้เหลือเฉพาะสถานีที่ผ่านเกณฑ์ (ข้อมูลหาย < 20%)
# ใช้ตัวแปร df_filtered ที่คุณรันผ่านแล้วมาทำต่อครับ
df_ready = df_filtered[~df_filtered['tele_station_oldcode'].isin(drop_list)].copy()

print(f"จำนวนสถานีที่ถูกคัดออก: {len(drop_list)} สถานี")
print(f"คงเหลือสถานีที่นำไปวิเคราะห์ต่อ: {df_ready['tele_station_oldcode'].nunique()} สถานี")
```

```python
# --- ส่วนเพิ่มเติมของข้อ 4: แสดงรายชื่อสถานีที่ถูกคัดออก ---

# 1. ดึงรายชื่อสถานีที่อยู่ใน drop_list มาแสดง
print(f"--- รายชื่อสถานีที่ถูกคัดออกทั้งหมด ({len(drop_list)} สถานี) ---")
print("-" * 50)

# แสดงรายชื่อสถานี แบ่งกลุ่มละ 10 ชื่อเพื่อให้ตรวจสอบง่าย
for i in range(0, len(drop_list), 10):
    print(", ".join(drop_list[i:i+10]))

print("-" * 50)

# 2. แสดงตารางสรุปสถานีที่ถูกคัดออกพร้อมเปอร์เซ็นต์ข้อมูลที่หายไป (เพื่อเป็นหลักฐานยืนยัน)
removed_summary = summary[summary['missing_ratio'] >= 20].sort_values('missing_ratio', ascending=False)
display(removed_summary[['Node', 'actual_days', 'missing_ratio']])
```
